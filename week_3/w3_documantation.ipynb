# 1: Identify Components
### 1. Identify the Inputs
    quiz1: The mark for the first quiz (numeric/float).

    quiz2: The mark for the second quiz (numeric/float).

    quiz3: The mark for the third quiz (numeric/float).

another: The user's choice string (yes/no) to determine whether to continue the loop.

### 2. What is the Process?
    Looping: Keep repeating the process until the user decides to stop.

    Calculation: Calculate the average by adding the three quiz marks together and dividing the sum by 3.

    Evaluation: Compare the average against the passing threshold (50). If the average is 50 or greater, assign a PASS status; otherwise, assign a FAIL status.

    Termination Check: Ask the user if they want to check another student. If they type anything other than 'yes' or 'y', exit the loop.

### 3. What is the Output?
    The calculated Average Mark (formatted to 2 decimal places).

    The Result status ("PASS!" or "Result: FAIL!").

    A final goodbye message ("Goodbye!") when exiting.

# 2: Design the Algorithm
### 2.1 Pseudocode
BEGIN
    WHILE True DO
        PROMPT user to enter quiz1, quiz2, quiz3
        CONVERT inputs to floats
        
        COMPUTE average = (quiz1 + quiz2 + quiz3) / 3
        DISPLAY average
        
        IF average >= 50 THEN
            DISPLAY "Result: PASS!"
        ELSE
            DISPLAY "Result: FAIL!"
        ENDIF
        
        PROMPT user "Do you want to enter marks for another student? (yes/no): "
        READ another
        
        IF lowercase(another) is NOT "yes" AND lowercase(another) is NOT "y" THEN
            DISPLAY "Goodbye!"
            BREAK loop
        ENDIF
    ENDWHILE
END

### 2.2 Identify the Algorithm Structure
This algorithm utilizes all three fundamental programming structures:

Iteration (Looping): The entire program runs inside a while True loop, allowing multiple student scores to be processed sequentially until broken by the user.

Sequence: Inside the loop, the input collection, average calculation, and initial printing happen line-by-line, one after another in order.

Selection (Decision): * An if-else statement is used to branch the outcome into either a PASS or FAIL response depending on the average.

A second if statement decides whether to break out of the iteration or repeat it based on the user's input.
